```python
loaded_model = keras.models.load_model('/content/CNN_3.h5')
```

```python
test_image = image.load_img(r'/content/cat_0_0.jpg',target_size=(64,64))


#Data Security :

# ******************-------------------------------------------------------******************

#Encryption :

try:
	# take path of image as a input
	path = input(r' path of the Image : ')
	
	# taking encryption key as input
	key = int(input('Enter Key for encryption of Image : '))
	
	# print path of image file and encryption key that
	# we are using
	print(' Path of file : ', path)
	print('Key for encryption : ', key)
	
	# open file for reading purpose
	x = open(path, 'rb')
	
	# storing image data in variable "img"
	img = x.read()
	x.close()
	
	# converting image into byte array
	img = bytearray(img)


	# performing XOR operation on each value of bytearray
	for index, values in enumerate(img):
		img[index] = values ^ key

	# opening file for writing purpose
	x = open(path, 'wb')
	
	# writing encrypted data in image
	x.write(img)
	x.close()
	print('Encryption Completed')

	#Exception case
except Exception:
	print('Error caught,Try with numerical Key : ', Exception._name_)
 
 # *************************-----------------------------------------------***************************

plt.imshow(test_image)
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
result = loaded_model.predict(test_image)
# print(result)
if result[0][0] >= 0.5:
    p = 'normal'
else:
    p = 'cataract'
print(p)
```

```python
#Decryption

try:
	
	path = input(r'Enter path of Image : ')

	key = int(input('Enter Key for encryption of Image : '))
	
	print('The path of file : ', path)

	print('Key for Decryption : ', key)
	
	X = open(path, 'rb')
	
	# storing image data in variable "image"
	img = X.read()
	X.close()
	
	img = bytearray(img)

	# performing XOR operation on each value of bytearray
	for index, values in enumerate(img):
		img[index] = values ^ key

	X = open(path, 'wb')
	
	X.write(img)
	X.close()
	print('Decryption Done...')

 #Exception case
except Exception:
	print('Error caught, Wrong key : ', Exception._name_)
```
